```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ds = pd.read_csv('all_stocks_5yr.csv')
```

```python
ds['close'].plot(figsize=(10,8))
```

```python
df = ds.loc[range=(100,1100)]
```

```python
df.loc[range(200,800)].rolling(12).std().plot(label='std of data',figsize=(8,6))
df.loc[range(200,800)].rolling(12).mean().plot(label='mean of data',figsize=(8,6))
df.loc[range(200,800)].plot(label='actual data' ,figsize=(8,6))
plt.legend()
```

```python
import statsmodels.api as sta
from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(df,freq=12, model='additive')
decomp.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df)
result
```
